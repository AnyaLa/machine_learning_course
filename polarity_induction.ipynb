```python
root_embed = '/home/nst/mount/data/linguistics_hse/machine_learning/hw_3/wiki.ru.vec'
```

```python
words = []

embeddings_index = {}
fo = open(root_embed)
for line in tqdm(fo):
    values = line.split()
    if len(values) == 301:
        word = values[0]
        words.append(word)
        coefs = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = coefs
fo.close()
```

```python
print(embeddings_index['красивого'])
```

```python
books_train = pd.read_csv('/home/nst/mount/data/share/dropbox/Dropbox/ML_sent/books_train.tsv', sep='\t', index_col=0)
```

```python
books_train = books_train[:500]
```

```python
tokenizer_books = Tokenizer()
tokenizer_books.fit_on_texts(books_train.lemmatized)
```

```python
word_index = tokenizer_books.word_index
```

```python
len(word_index)
```

```python
vocabular = list(word_index.keys())
```

```python
#какая строчка из матрицы соответствует каждому слову из словаря
embedding_matrix = np.zeros((len(word_index), 300)) 
for word, i in word_index.items():
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None:
        # words not found in embedding index will be all-zeros.
        embedding_matrix[i-1] = embedding_vector
```

```python
embedding_matrix.shape
```

```python
embeddings = embedding.Embedding(embedding_matrix, vocabular, normalize=True)
```

```python
POSITIVE_BOOKS = ['понравиться','интересный', 'великолепный', 'обожать', 'приятный', 'захватывать', 'замечательный', 'отличный', 'любимый', 'супер', 'обалденный', 'восторг']
NEGATIVE_BOOKS = ['разочарование', 'плохой', 'банальность', 'скомканный', 'ужасный', 'пресытиться', 'худой', 'нудно', 'извращать', 'средний', 'облом', 'мусорный']
```

```python
polarties = polarity_induction_methods.random_walk(embeddings, POSITIVE_BOOKS, NEGATIVE_BOOKS)
```

```python
from socialsent import seeds
from socialsent import lexicons
from socialsent.polarity_induction_methods import label_propagate_continuous
from socialsent.evaluate_methods import binary_metrics
from socialsent.representations.representation_factory import create_representation
```

```python
embed_root = '/home/nst/mount/data/linguistics_hse/machine_learning/hw_3/wiki.ru.vec'
```

```python
lexicon = lexicons.load_lexicon("rus_lexicon", remove_neutral=True)
pos_seeds, neg_seeds = seeds.book_seeds()
```

```python
eval_words = [word for word in embeddings if not word in pos_seeds 
              and not word in neg_seeds]
```

```python
polarities_books = polarity_induction_methods.random_walk(embeddings, pos_seeds, neg_seeds, beta=0.99, nn=10, sym=True, arccos=True)
```

```python
acc, auc, avg_per  = binary_metrics(polarities_books, lexicon, eval_words)
print "Accuracy with best threshold: {:0.2f}".format(acc)
print "ROC AUC: {:0.2f}".format(auc)
print "Average precision score: {:0.2f}".format(avg_per)
```
